```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


input_words = ["I", "love", "deep", "learning"]
output_words = ["J'", "aime", "l'apprentissage"]

attention_weights = np.array([
    [0.7, 0.2, 0.05, 0.05],
    [0.1, 0.8, 0.05, 0.05],
    [0.05, 0.1, 0.4, 0.45]
])

plt.figure(figsize=(8,4))

sns.heatmap(
    attention_weights,
    annot=True,
    cmap="Blues",
    xticklabels=input_words,
    yticklabels=output_words
)

plt.xlabel("Input Sequence")
plt.ylabel("Output Sequence")
plt.title("Attention Heatmap in Seq2Seq")

plt.show()
```
